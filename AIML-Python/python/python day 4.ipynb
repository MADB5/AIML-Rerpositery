```python
try:
    units = float(input("Enter the number of units consumed: "))

    if units <= 50:
        bill = units * 0.50
    elif units <= 150:
        bill = 50 * 0.50 + (units - 50) * 0.75
    elif units <= 250:
        bill = 50 * 0.50 + 100 * 0.75 + (units - 150) * 1.20
    else:
        bill = 50 * 0.50 + 100 * 0.75 + 100 * 1.20 + (units - 250) * 1.50

    total_bill = bill + (bill * 0.20)  # Adding 20% surcharge

    print("Total Electricity Bill:", total_bill)
    
    try:
        if units < 0 :
          raise Exception("Invalid units.")
    except Exception as e:
        print(e)
        sys.exit(1)
        
except ValueError:
    print("Invalid input! Please enter numeric values for units consumed.")
```

```python
# Wordcount.py

import sys

def word_count(filename):
    word_counts = {}
    
    with open(filename, 'r') as file:
        for line in file:
            words = line.strip().split()
            for word in words:
                if word in word_counts:
                    word_counts[word] += 1
                else:
                    word_counts[word] = 1
    
    return word_counts

if __name__ == '__main__':
    if len(sys.argv) != 2:
        print("Usage: python wordcount.py <filename>")
        sys.exit(1)
    
    filename = sys.argv[1]
    counts = word_count(filename)
    
    for word, count in counts.items():
        print(f"{word}: {count}")
        
# Commend.py

import sys
from wordcount import word_count

if __name__ == '__main__':
    if len(sys.argv) != 2:
        print("Usage: python command.py <filename>")
        sys.exit(1)
    
    filename = sys.argv[1]
    counts = word_count(filename)
    
    for word, count in counts.items():
        print(f"{word}: {count}")
        
      '''
        in prompt we call commend.py with a text file that 
        has words to count that is apple.txt
        # python commend.py apple.txt
        
        this will exhigute the commend.py and wordcount
        and prints the output with word and its count
        
        OUTPUT
        
hello: 2
rajuhello: 1
hi: 1
wonder: 1
way: 1
        '''
```

# 5. Write a Python program for finding the most frequent words in a text read from a file. 
Initially open the text file in read mode. 

Make all the letters in the document into lowercase letters and split the words in cach line.

Get the words in an order.

Sort the words for finding the most frequent words in the file.

Print the most frequent words in the file

```python
from collections import Counter

def find_most_frequent_words(filename):
    word_counts = Counter()
    
    with open(filename, 'r') as file:
        for line in file:
            words = line.lower().split()
            word_counts.update(words)
    
    ordered_words = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)
    
    return ordered_words

if __name__ == '__main__':
    filename = "d4prog5.txt"
    
    frequent_words = find_most_frequent_words(filename)
    
    print("The most frequent words in the file are:")
    for word, count in frequent_words:
        print(f"{word}: {count}")
        
        '''
        in prompt we call d4prog5p.py with a text file that 
        has words to count the most frequent words that is d4prog5.txt
        # python d5prog4p.py d4prog5.txt
        
        this will exhigute the d5prog4p.py and checks for frequent words
        and prints the output with word and its frequent
        
        OUTPUT
        
women: 2
hi: 1
hello: 1
hay: 1
man: 1
king: 1
queen: 1
curse: 1
ferious: 1
wonder: 1

'''
```
